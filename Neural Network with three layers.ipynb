```python
import numpy as np
np.random.seed(1)

#Input array
X=np.array([[1,0,1,0],[1,0,1,1],[0,1,0,1]])

#Output
y=np.array([[1],[1],[0]])

#Sigmoid Function
#here we use numpy . because math package return a single value
def sigmoid (x):
    return 1/(1 + np.exp(-x))

#Derivative of Sigmoid Function
def derivatives_sigmoid(x):
    return x * (1 - x)

def initial_weights(n0,n1,n2,n3):
    W1=np.random.uniform(size=(n0,n1))
    b1=np.random.uniform(size=(1,n1))

    W2=np.random.uniform(size=(n1,n2))
    b2=np.random.uniform(size=(1,n2))

    W3=np.random.uniform(size=(n2,n3))
    b3=np.random.uniform(size=(1,n3))
    return W1,b1,W2,b2,W3,b3
    
def forward_propagation(input_,W,B):
    
    temp_=np.dot(input_,W) + B
    return sigmoid(temp_)
    

#Variable initialization
epoch=10000 # number of iterations
lr=0.05 # learning rate 
n0 = X.shape[1] #dimention of input . here it is 4
n1 = 3 #number of  neurons at layer 2
n2 = 3 #number of neurons at layer 3
n3 = 1 #number of neurons at output layer
#weight and bias initialization
# set intial value to W1 and b1
W1,b1,W2,b2,W3,b3=initial_weights(n0,n1,n2,n3)

for i in range(epoch):

#calculate forward_propagation    
    a1=forward_propagation(X,W1,b1)
    a2=forward_propagation(a1,W2,b2)
    output=forward_propagation(a2,W3, b3)
    
    
    

    #Backpropagation
    dz3= (y-output) * derivatives_sigmoid(output)
    #here we calculate new value for matrix W3 and constant b3
    W3 += a2.T.dot(dz3) *lr
    b3 += np.sum(dz3, axis=0,keepdims=True) *lr
    
    dz2 =  dz3.dot(W3.T)* derivatives_sigmoid(a2)
    
    #here we calculate new value for matrix W2 and constant b2
    # we calculate the amount of changes and update W2,b2
    W2 += a1.T.dot(dz2) *lr
    b2 += np.sum(dz2, axis=0,keepdims=True) *lr   
      
    dz1 =  dz2.dot(W2.T)* derivatives_sigmoid(a1)
    
    #here we calculate new value for matrix W1 and constant b1
    # we calculate the amount of changes and update W1,b1    
    W1 += X.T.dot(dz1) *lr
    b1 += np.sum(dz1, axis=0,keepdims=True) *lr
    if (i% 1000 ==0 ):
        #average of Cross-entropy error function 
        loss = -np.mean ( y * np.log(output) + (1 - y) * np.log(1 - output) )
        print (loss)

print (output)
loss = -np.mean ( y * np.log(output) + (1 - y) * np.log(1 - output) )
print (loss)
```
